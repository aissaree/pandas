```python
import pandas as pd
df=pd.read_excel('Employee.xlsx',index_col='Name')
df
```

```python
# df.dropna(inplace=True) # แถวใดมีข้อมูลว่าง จะทำการ Drop
# df
```

```python
#ลบแถวบางส่วนที่มีค่าว่าง
df.dropna(subset=['Job','Salary'],inplace=True)
df
```

```python
#ลบ column ที่ข้อมูลไม่ครบ
df.dropna(axis='columns',inplace=True)
df
```
