```python
import pandas as pd
df=pd.read_excel('Employee.xlsx',index_col='Name')
df
```

```python
df.isnull() #แสดงผลเป็นค่าทางตรรกศาสตร์ True คือเป็น null
```

```python
df.isnull().any() #column ที่มีค่าว่าง จะแสดงค่าเป็น True
```

```python
df.isnull().sum() #แสดงผลว่าแต่ละ column มีค่าว่างกี่แถว
```

```python
df.notnull() #ไม่ว่างจะแสดง True
```

```python
df.notnull().sum()
```
